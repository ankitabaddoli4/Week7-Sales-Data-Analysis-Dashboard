```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Ankita Baddoli\notebooks\data\processed\clean_sales_data.csv')
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

```python
print("Total Sales:", df['total_amount'].sum())
print("Average Sales:", df['total_amount'].mean())
```

```python
category_sales = df.groupby('category')['total_amount'].sum()
category_sales
```

```python
df['month'] = df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('month')['total_amount'].sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', marker='o', title='Monthly Sales')
plt.show()
```

```python
category_sales.plot(kind='bar', title='Category Sales')
plt.show()
```

```python
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
import os
os.makedirs('data/reports', exist_ok=True)

category_sales.to_csv('data/reports/category_sales.csv')
monthly_sales.to_csv('data/reports/monthly_sales.csv')

print("✅ Reports saved")
```
