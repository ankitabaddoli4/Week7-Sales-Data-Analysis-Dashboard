```python
import os
import pandas as pd

# 🔹 Step 1: Create folders
os.makedirs('data/raw', exist_ok=True)
os.makedirs('data/processed', exist_ok=True)

# 🔹 Step 2: Create dataset
data = {
    "order_id": [1,2,3,4,5,6,7],
    "order_date": ["2024-01-05","2024-01-10","2024-02-15","2024-03-12","2024-03-18","2024-04-20","2024-05-25"],
    "customer_id": ["C001","C002","C001","C003","C002","C004","C005"],
    "product_id": ["P001","P002","P003","P004","P002","P005","P006"],
    "category": ["Electronics","Clothing","Electronics","Books","Clothing","Sports","Home"],
    "quantity": [2,1,1,3,2,1,4],
    "total_amount": [500,200,300,150,400,250,600]
}

df = pd.DataFrame(data)

# 🔹 Step 3: Save RAW file
df.to_csv('data/raw/sales_data.csv', index=False)

# 🔹 Step 4: Clean data
df['order_date'] = pd.to_datetime(df['order_date'])
df = df.drop_duplicates()

# 🔹 Step 5: Save CLEAN file ⭐
df.to_csv('data/processed/clean_sales_data.csv', index=False)

print("✅ FILE CREATED SUCCESSFULLY")
```
